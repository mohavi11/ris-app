## مرحله ۶: دانلود مستقیم APK
این سلول رو اجرا کن تا فایل خودکار دانلود بشه (روی گوشی هم کار می‌کنه).

```python
import glob
from google.colab import files

apk_files = glob.glob('bin/*.apk')
if apk_files:
    print('در حال دانلود:', apk_files[0])
    files.download(apk_files[0])
else:
    print('هیچ فایل APK پیدا نشد. لاگ بیلد رو بررسی کن.')
```
